```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Datos proporcionados
area = [76, 92, 107, 42, 65, 150, 65, 105, 40, 55]
valor = [200, 320, 300, 160, 190, 315, 230, 380, 155, 175]
```

```python
# Crear un DataFrame con los datos
df = pd.DataFrame({'area': area, 'valor': valor})
```

```python
# Visualizar los datos
plt.scatter(df['area'], df['valor'])
plt.xlabel('Área')
plt.ylabel('Valor')
plt.title('Gráfico de dispersión: Área vs. Valor')
plt.show()
```

```python
# Realizar la regresión lineal
modelo = LinearRegression()
X = df[['area']]
y = df['valor']
modelo.fit(X, y)
```

```python
# Calcular los valores predichos por el modelo
y_pred = modelo.predict(X)
```

```python
# Trazar la línea de regresión junto con los puntos de datos
plt.scatter(df['area'], df['valor'])
plt.plot(df['area'], y_pred, color='red', label='Regresión Lineal')
plt.xlabel('Área')
plt.ylabel('Valor')
plt.title('Regresión Lineal: Área vs. Valor')
plt.legend()
plt.show()
```
